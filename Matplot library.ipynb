# Matplotlib 

* Matplotlib is a multi-platform data visualization library built on NumPy arrays.
* Matplot library is mother of all library.
* One of the greatest benefits of visualization is that it allows us visual access to huge amounts of data in easily digestible   visuals.
* Matplotlib consists of several plots like line, bar, scatter, histogram many more.
* It is a easy to understand.

###  Pyplot

* Most of the matplotlib utilities lies under the pyplot submodule, and are usually imported under plt alias

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import matplotlib as mlt
plt.style.use("default")
```

# Line Plot

```python
# create to numpy arrays and use plt.plot to create a line chart
x =np.array([0,6])                  # we can also use np.arrange(0,6)         
print(x)
y= np.array([0,250])
print(y)
```

```python
# create to numpy arrays and use plt.plot to create a line chart
plt.plot(x,y)
plt.title('Line Plot')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.show()
```

```python
# we can use some filter and change line colour and style
plt.plot(x,y, color ='b',linestyle = ":",linewidth= 2 )
plt.title("Line plot")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.
```

```python
x = np.array([12,3,4,45,33,66])
# mec parameter is used to set the color for edges of marker
# mec = marker edges color
# mfc = marker face color
# mfc is used to set the color for the face of the marker
plt.plot(x,"D:m",ms = 15, mec = "g",mfc ="k")
plt.show()
```

```python
x=np.arange(0,11)
y=np.arange(11,21)
y =x*x
plt.plot(x,y)
plt.title("Title")
plt.xlabel("X-label")
plt.ylabel("Y-label")
plt.show()
```

# Subplot

```python
x= np.arange(10,30)
y= np.arange(30,50)
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

# Bar Plot

```python
x = np.array(["A","B","C","D"])
y = np.array([5,8,2,13])
plt.bar(x,y)
plt.show()
```

# Horizontal Bar Plot

```python
x = np.array(["A","B","C","D"])
y = np.array([3,8,1,10])
plt.barh(x,y,color='Orange')
plt.show()
```

# Piechart

```python
x = np.random.randint(1, 100, size=5)
mylabels = ['Book', 'Pen', 'Board', 'Bench' ,'Pencik']
plt.pie(x, labels=mylabels, autopct = "%1.1f%%")
plt.show()
```

```python
Fruit = ['Apple', 'Orange', 'Mango', 'Guava']
quantity = [67, 43, 100, 29]
plt.pie(quantity, labels=Fruit, autopct='%1.1f%%', radius=2)
plt.pie([1], colors=['w'], radius=1)
plt.axis('equal')  
plt.title('Donut Chart - Fruit Distribution')
plt.show()
```

# KDE Plot

```python
iris = sns.load_dataset('iris')
#histogram
sns.histplot(iris['sepal_width'],bins =25,kde = True)   #kde give smooth curve to represent actule value
plt.show()
```

# Contour plot
